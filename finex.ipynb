```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
df1 = data
df1 = df1.rename(columns={'whoAmI': 'ROBOT'})
for i in range(0, len(df1)):
  if (df1['ROBOT'][i] != 'robot'):
    df1['ROBOT'][i] = 0
  else:
    df1['ROBOT'][i] = 1
df1
```

```python
df2 = data
df2 = df2.rename(columns={'whoAmI': 'HUMAN'})
for i in range(0, len(df2)):
  if (df2['HUMAN'][i] != 'human'):
    df2['HUMAN'][i] = 0
  else:
    df2['HUMAN'][i] = 1
df2
```

```python
df3 = pd.concat([df1,df2], axis=1)
df3
```
